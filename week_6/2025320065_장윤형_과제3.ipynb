```
# 필요한 패키지 import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 실습용 데이터 파일 생성
iris = sns.load_dataset('iris')
iris.to_excel('iris.xlsx')

flights = sns.load_dataset('flights')
flights.to_excel('flights.xlsx')

tips = sns.load_dataset('tips')
tips.to_excel('tips.xlsx')

titanic = sns.load_dataset('titanic')
titanic.to_excel('titanic.xlsx')
```

> 연도별 비행기 탑승자 분석



```
# flights 연도별 비행기 탑승객 변화 분석
# 엑셀 데이터 load 및 확인
flights=pd.read_excel('flights.xlsx')
flights
```

```
# flights 데이터의 피봇 테이블 생성하기
flights_passenger = flights.pivot(index='month', columns='year',  values= 'passengers')
flights_passenger
```

```
# 월별로 다시 정렬하기
flights_passenger = flights_passenger.reindex(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
flights_passenger
```

문제 6

```
#todo 6 flights 데이터의 시각화, line 3 작성
# seaborn 의 heatmap함수를 적용하여 아래 출력결과와 동일하게(축설정, 색상, 형식 등) 출력되도록 코딩
sns.heatmap(data=flights_passenger, annot=True, fmt='d')
```
